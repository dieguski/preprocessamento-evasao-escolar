## 2. Pré-processamento

### 2.1 Limpeza de dados:

#### 2.1.1 Registros duplicados:

Vamos verificar se há linhas duplicadas

```python
duplicados = df_new.duplicated(keep='first')

print(f'Linhas duplicadas: {duplicados.sum()}')

df_new[duplicados]
```

Visualizando ambas as linhas repetidas:

```python
df_new[df_new.duplicated(keep=False)]
```

Excluindo os primeiros registros duplicados e verificando se o procedimento foi realizado com sucesso: 

```python
df_sem_duplicadas = df_new.drop_duplicates().copy()
print(f'Linhas duplicadas: {df_sem_duplicadas.duplicated().sum()}')
```

#### 2.1.2 Valores nulos:

Primeiramente realizamos uma verificação da quantidade total de linhas do DF completo através da função '.shape', e em seguida apresentamos a quantidade de valores nulos em cada coluna.

Obs.: tentei utilizar o .describe() da primeira vez, mas notei que ele não trazia as infos de colunas não numéricas. Deste modo, optei por contar os valores nulos em cada coluna.

```python
print(f'Shape do DF sem duplicadas: {df_sem_duplicadas.shape[0]}')

nulos = df_sem_duplicadas.isna().sum()
print(f'\nQuantidade de valores nulos em cada coluna:\n{nulos}')
```

Notamos que as colunas 'Renda Familiar ($)', 'Escola Ensino Médio', 'Trabalha Atualmente' e 'Desempenho' possuem a mesma de quantidade de valores nulos: 8.

Outro ponto interessante é que a coluna 'Motivo da Evasão' apresenta 101 valores nulos (alunos que estão ativos ou se formaram), ou seja, 49 alunos que evadiram (150 - 101 = 49).

Para validar se esse número realmente representa a quantidade de alunos evadidos e garantir que não haja valores não preenchidos, realizamos uma verificação da quantidade de alunos evadidos baseado no shape do DF:

```python
evasao = df_sem_duplicadas[df_sem_duplicadas['Status'] == 'Evadido']
print(f'\nShape da quantidade de alunos evadidos: {evasao.shape[0]}')
```

Notamos uma discrepância entre os valores, onde anteriormente tínhamos o valor de 49 alunos evadidos, agora percebemos que há na verdade 54 alunos evadidos.

Isso indica que há 5 alunos evadidos com a coluna 'Motivo da Evasão' não preenchida.

Vamos validar filtrando pelos alunos evadidos e com a coluna em questão com valores nulos:

```python
evadidos_sem_motivo = df_sem_duplicadas[(df_sem_duplicadas['Status'] == 'Evadido') & (df_sem_duplicadas['Motivo da Evasão'].isna())]
print(f'\nQuantidade de registros evadidos sem motivo de evasão (nulo): {evadidos_sem_motivo.shape[0]}')
```

Vamos substituir esses valores nulos por 'Não Informado' para não introduzir viés em nosso DF, e em seguida validar se os valores nulos para o status evadido foram zerados.

```python
filtro = (df_sem_duplicadas['Status'] == 'Evadido') & (df_sem_duplicadas['Motivo da Evasão'].isna())

df_sem_duplicadas.loc[filtro, 'Motivo da Evasão'] = 'Não Informado'

evadidos_sem_motivo = df_sem_duplicadas[(df_sem_duplicadas['Status'] == 'Evadido') & (df_sem_duplicadas['Motivo da Evasão'].isna())]
print(f'\nQuantidade de registros evadidos sem motivo de evasão (nulo): {evadidos_sem_motivo.shape[0]}')
```

Vamos agora verificar as outras colunas com valores nulos:

```python
colunas_nulo = df_sem_duplicadas[
    (df_sem_duplicadas['Renda Familiar (R$)'].isna()) +
    (df_sem_duplicadas['Escola Ensino Médio'].isna()) +
    (df_sem_duplicadas['Trabalha Atualmente'].isna()) +
    (df_sem_duplicadas['Desempenho'].isna())]

qtd_linhas = colunas_nulo.shape[0]
porcentagem = int(colunas_nulo.shape[0]) / int(df_sem_duplicadas.shape[0]) * 100

print(f'Quantidade de linhas onde alguma das colunas\n(Renda Familiar (R$),\nEscola Ensino Médio,\nTrabalha Atualmente,\nDesempenho)\ntem valor nulo: {qtd_linhas}')
print(f'Porcentagem de {qtd_linhas} linhas em relação ao DF completo: {porcentagem}')

colunas_nulo
```

29 linhas e 19,33% do DF completo é bastante coisa, então não podemos remover as linhas. Vamos substituir os valores. Podemos notar na coluna Desempenho que o valor máximo está sendo apresentado como 150, o que indica um outlier e um erro, tendo em vista que o desempenho é baseado em valores de 0 a 100.

```python
df_sem_duplicadas['Desempenho'].describe()
```

Deste modo, vamos calcular o valor da mediana sem considerar esse valor de 150, transformar esse valor em nulo, em seguida imputar o valor da mediana nessa posição, e então validar se está devidamente ajustado.

```python
df_sem_duplicadas.loc[df_sem_duplicadas['Desempenho'] > 100, 'Desempenho'] = None

mediana_desempenho = df_sem_duplicadas['Desempenho'].median()

df_sem_duplicadas['Desempenho'] = df_sem_duplicadas['Desempenho'].fillna(mediana_desempenho)

df_sem_duplicadas.describe()
```

Show de bola! Vamos para a próxima coluna: Renda Familiar (R$).

Nesta coluna, vamos realizar o mesmo procedimento realizado na coluna Desempenho, tendo em vista o valor outlier de 50000.

Com isso, vamos calcular o valor da mediana sem considerar esse valor de 500000, transformar esse valor em nulo, em seguida imputar o valor da mediana nessa posição, e então validar se está devidamente ajustado.

```python
print(df_sem_duplicadas['Renda Familiar (R$)'].sort_values(ascending=False))
```

```python
df_sem_duplicadas.loc[df_sem_duplicadas['Renda Familiar (R$)'] > 4900, 'Renda Familiar (R$)'] = None

mediana_renda = df_sem_duplicadas['Renda Familiar (R$)'].median()

df_sem_duplicadas['Renda Familiar (R$)'] = df_sem_duplicadas['Renda Familiar (R$)'].fillna(mediana_renda)

df_sem_duplicadas['Renda Familiar (R$)'].describe()
```

Vamos para a próxima coluna: Escola Ensino Médio.

Nesta situação, para sermos conservadores, vamos alterar os valores nulos para "Desconhecida".

```python
df_sem_duplicadas['Escola Ensino Médio'] = df_sem_duplicadas['Escola Ensino Médio'].fillna('Desconhecida')

print(df_sem_duplicadas['Escola Ensino Médio'].isna().sum())
```

Última coluna a ter os valores nulos tratados: Trabalha Atualmente

Vamos realizar o mesmo processo da coluna 'Escola Ensino Médio', imputando 'Desconhecido' aos valores nulos.

```python
df_sem_duplicadas['Trabalha Atualmente'] = df_sem_duplicadas['Trabalha Atualmente'].fillna('Desconhecido')

print(df_sem_duplicadas['Trabalha Atualmente'].isna().sum())
```

Podemos observar que agora não temos mais valores nulos (desconsiderando os 101 registros nulos de Motivos da Evasão, que representam alunos ativos ou formados):

```python
df_tratado = df_sem_duplicadas

df_tratado.isna().sum()
```

### 2.4 Transformação de dados:

#### 2.4.1 Padronizando valores de texto:

Ao visualizar as primeiras linhas do DF, podemos observar uma inconsistência no padrão dos valores da coluna 'Gênero':

```python
df_tratado.head()
```

Vamos validar em quais colunas a mesma inconsistência ocorre:

Notamos que as colunas 'Gênero' e 'Estado Civil' possuem mais de uma palavra para um mesmo dado. Com isso, vamos padronizar e organizar estes dados, validando se os ajustes ocorreram corretamente.

```python
for col in df_tratado.columns:
    print(df_tratado[col].value_counts())
    print()
```

```python
#Gênero
#Feminino     57
#Masculino    47
#Outro        43
#feminino      1
#FEM           1
#masc          1

df_tratado['Gênero'] = df_tratado['Gênero'].replace(['feminino', 'FEM'], 'Feminino')
df_tratado['Gênero'] = df_tratado['Gênero'].replace('masc', 'Masculino')

#Estado Civil
#Casado(a)        46
#Viúvo(a)         39
#Divorciado(a)    34
#Solteiro(a)      27
#Solteir           1
#Casada            1
#Viuvo             1
#Separado          1

df_tratado['Estado Civil'] = df_tratado['Estado Civil'].replace('Solteir', 'Solteiro(a)')
df_tratado['Estado Civil'] = df_tratado['Estado Civil'].replace('Casada', 'Casado(a)')
df_tratado['Estado Civil'] = df_tratado['Estado Civil'].replace('Viuvo', 'Viúvo(a)')
df_tratado['Estado Civil'] = df_tratado['Estado Civil'].replace('Separado', 'Divorciado(a)')


print(df_tratado['Gênero'].value_counts())
print()
print(df_tratado['Estado Civil'].value_counts())
```

#### 2.4.2 Conversão de variáveis categóricas para números

```python
mapeamento = {
    'Gênero':               {'Feminino': 1, 'Masculino': 2, 'Outro': 3},
    'Estado Civil':         {'Solteiro(a)': 1, 'Divorciado(a)': 2, 'Viúvo(a)': 3, 'Casado(a)': 4},
    'Escola Ensino Médio':  {'Pública': 1, 'Privada': 2, 'Desconhecida': 3},
    'Status':               {'Ativo': 1, 'Evadido': 2, 'Formado': 3},
    'Trabalha Atualmente':  {'Não': 1, 'Sim': 2, 'Desconhecido': 3} 
}

for col, mapa in mapeamento.items():
    df_tratado[col] = df_tratado[col].map(mapa)

df_tratado.head()
```

### 2.4.3 Criação de variáveis derivadas

```python
for i in df_tratado.index:
    idade = df_tratado.loc[i, 'Idade']

    if idade <= 21:
        df_tratado.loc[i, 'Faixa Etária'] = 'Jovem'
    elif idade <= 60:
        df_tratado.loc[i, 'Faixa Etária'] = 'Adulto'
    else:
        df_tratado.loc[i, 'Faixa Etária'] = 'Maduro'
```

```python
min_renda = df_tratado['Renda Familiar (R$)'].min()
max_renda = df_tratado['Renda Familiar (R$)'].max()

print(min_renda, max_renda)

df_tratado['Renda Familiar Normalizada'] = (df_tratado['Renda Familiar (R$)'] - min_renda) / (max_renda - min_renda)
```
