Carregando dataset e visualizando informações básicas para entender sua estrutura.

### 1.1 Valores nulos e coluna duplicada

A partir dessa análise inicial, notamos que:
1. a coluna 'Renda Familiar Total' é duplicada em relação à coluna 'Renda Familiar (R$)';
2. as colunas 'Renda Familiar (R$)', 'Escola Ensino Médio', 'Motivo da Evasão', 'Trabalha Atualmente' e 'Desempenho' possuem valores nulos - Tendo em vista que o dataframe possue 153 linhas;

Filtrando dataframe apenas para as colunas que iremos utilizar, removendo a coluna duplicada (Renda Familiar Total).

```python
df = df[['Idade', 'Gênero', 'Estado Civil', 'Renda Familiar (R$)', 'Escola Ensino Médio', 'Status', 'Motivo da Evasão', 'Trabalha Atualmente', 'Último Período Cursado', 'Desempenho']].copy()
```

## 2. Pré-processamento

### 2.1 Limpeza de dados

#### 2.1.1 Registros/linhas duplicados

##### 2.1.1.1 Verificando se há linhas duplicadas no DF

```python
duplicados = df.duplicated(keep=False)

print(f'Linhas duplicadas: {duplicados.sum()}')

df[duplicados]
```

Notamos repetição entre as linhas dos índices abaixo:
- 14 e 150
- 75 e 152
- 98 e 151

Assim, será necessário remover 3 linhas, evitando a duplicidade.

##### 2.1.1.2 Excluindo registros duplicados e verificando sucesso do ajuste: 

```python
df = df.drop_duplicates().copy()
print(f'Linhas duplicadas: {df.duplicated().sum()}')
```

#### 2.1.2 Valores nulos:

Nas seções 1 e 1.1 notamos que o dataframe possui algumas colunas com valores nulos. Vamos investigar esses valores nulos mais a fundo. 

```python
print(f'Shape do DF sem duplicadas: {df.shape[0]}')

nulos = df.isna().sum()
print(f'\nQuantidade de valores nulos em cada coluna:\n{nulos}')
```

* A coluna 'Motivo da Evasão' apresenta 101 valores nulos (alunos que estão ativos ou se formaram), ou seja, 49 alunos que evadiram (150 - 101 = 49).

* Notamos que as colunas 'Renda Familiar ($)', 'Escola Ensino Médio', 'Trabalha Atualmente' e 'Desempenho' possuem a mesma de quantidade de valores nulos: 8.

##### 2.1.2.1 Validando evadidos com motivo de evasão nulo

Para validar se esse número realmente representa a quantidade de alunos evadidos e garantir que não haja valores não preenchidos, realizamos uma verificação da quantidade de alunos evadidos baseado no shape do DF:

```python
evasao = df[df['Status'] == 'Evadido']
print(f'Quantidade de alunos evadidos: {evasao.shape[0]}')
```

* Notamos uma discrepância entre os valores, onde anteriormente tínhamos o valor de 49 alunos evadidos, agora percebemos que há na verdade 54 alunos evadidos.

* Isso indica que há 5 alunos evadidos com a coluna 'Motivo da Evasão' não preenchida.

Vamos validar filtrando pelos alunos evadidos e com a coluna em questão com valores nulos:

```python
evadidos_sem_motivo = df[(df['Status'] == 'Evadido') & (df['Motivo da Evasão'].isna())]
print(f'Quantidade de registros evadidos sem motivo de evasão (nulo): {evadidos_sem_motivo.shape[0]}')
```

##### 2.1.2.2 Substituindo valores nulos de evadidos sem motivo

Vamos substituir esses valores nulos por 'Não Informado' para não introduzir viés em nosso DF, e em seguida validar se os valores nulos para o status evadido foram zerados.

```python
# Substituindo valores nulos por 'Não Informado'
filtro = (df['Status'] == 'Evadido') & (df['Motivo da Evasão'].isna())
df.loc[filtro, 'Motivo da Evasão'] = 'Não Informado'

# Verificando ajuste
evadidos_sem_motivo = df[(df['Status'] == 'Evadido') & (df['Motivo da Evasão'].isna())]
print(f'Quantidade de registros evadidos sem motivo de evasão (nulo): {evadidos_sem_motivo.shape[0]}')
```

##### 2.1.2.3 Substituindo valores nulos da coluna 'Desempenho'

Antes de preencher os valores nulos da coluna 'Desempenho', vamos calcular o valor da mediana sem considerar valores maiores que 100, transformar esse valor em nulo, em seguida imputar o valor da mediana nessa posição, e então validar se está devidamente ajustado.

```python
# Transformando valor de 150 em nulo 
df.loc[df['Desempenho'] > 100, 'Desempenho'] = None

# Definindo mediana e substituindo no df
mediana_desempenho = df['Desempenho'].median()
df['Desempenho'] = df['Desempenho'].fillna(mediana_desempenho)

df['Desempenho'].describe()
```

##### 2.1.2.4 Substituindo valores nulos da coluna 'Renda Familiar (R$)'

Nesta coluna, vamos realizar o mesmo procedimento realizado na coluna Desempenho, tendo em vista o valor outlier de 50000.

Com isso, vamos calcular o valor da mediana sem considerar esse valor de 500000, transformar esse valor em nulo, em seguida imputar o valor da mediana nessa posição, e então validar se está devidamente ajustado.

```python
# Visualizando as 5 maiores rendas
print(f'5 maiores rendas:\n{df['Renda Familiar (R$)'].sort_values(ascending=False).head()}')


# Transformando valor de 50000 em nulo 
df.loc[df['Renda Familiar (R$)'] > 4900, 'Renda Familiar (R$)'] = None


# Definindo mediana e substituindo no df
mediana_renda = df['Renda Familiar (R$)'].median()
df['Renda Familiar (R$)'] = df['Renda Familiar (R$)'].fillna(mediana_renda)

print('\nEstatísticas após ajuste:')
df['Renda Familiar (R$)'].describe()
```

##### 2.1.2.5 Substituindo valores nulos da coluna 'Escola Ensino Médio'.

Nesta situação, para sermos conservadores, vamos alterar os valores nulos para "Desconhecida" e verificar a quantidade de valores nulos para validar se o ajuste foi realizado com sucesso.

```python
df['Escola Ensino Médio'] = df['Escola Ensino Médio'].fillna('Desconhecida')

print(df['Escola Ensino Médio'].isna().sum())
```

##### 2.1.2.6 Substituindo valores nulos da coluna 'Trabalha Atualmente'

Vamos realizar o mesmo processo da coluna 'Escola Ensino Médio', imputando 'Desconhecido' aos valores nulos.

```python
df['Trabalha Atualmente'] = df['Trabalha Atualmente'].fillna('Desconhecido')

print(df['Trabalha Atualmente'].isna().sum())
```

##### 2.1.2.7 Validação de ajustes
Podemos observar que agora não temos mais valores nulos (desconsiderando os 96 registros nulos de Motivos da Evasão, que representam alunos ativos ou formados):

```python
df.isna().sum()
```

### 2.2 Transformação de dados
#### 2.2.1 Padronizando valores de texto
Ao visualizar as primeiras linhas do DF, podemos observar uma inconsistência no padrão dos valores da coluna 'Gênero':

```python
df.head()
```

Vamos validar em quais colunas a mesma inconsistência ocorre:

Notamos que as colunas 'Gênero' e 'Estado Civil' possuem mais de uma palavra para um mesmo dado. Com isso, vamos padronizar e organizar estes dados, validando se os ajustes ocorreram corretamente.

```python
for col in df.columns:
    print(df[col].value_counts())
    print()
```

```python
#Gênero
#Feminino     57
#Masculino    47
#Outro        43
#feminino      1
#FEM           1
#masc          1

df['Gênero'] = df['Gênero'].replace(['feminino', 'FEM'], 'Feminino')
df['Gênero'] = df['Gênero'].replace('masc', 'Masculino')

#Estado Civil
#Casado(a)        46
#Viúvo(a)         39
#Divorciado(a)    34
#Solteiro(a)      27
#Solteir           1
#Casada            1
#Viuvo             1
#Separado          1

df['Estado Civil'] = df['Estado Civil'].replace('Solteir', 'Solteiro(a)')
df['Estado Civil'] = df['Estado Civil'].replace('Casada', 'Casado(a)')
df['Estado Civil'] = df['Estado Civil'].replace('Viuvo', 'Viúvo(a)')
df['Estado Civil'] = df['Estado Civil'].replace('Separado', 'Divorciado(a)')


print(df['Gênero'].value_counts())
print()
print(df['Estado Civil'].value_counts())
```

#### 2.2.2 Criação de variáveis derivadas

```python
# Variáveis categóricas para idade:
df['Faixa Etária'] = pd.cut(df['Idade'],
                            bins = [0, 21, 49, 999],
                            labels = ['Jovem', 'Adulto', 'Maduro'])

# Normalizando renda:
min_renda = df['Renda Familiar (R$)'].min()
max_renda = df['Renda Familiar (R$)'].max()

df['Renda Familiar Normalizada'] = (df['Renda Familiar (R$)'] - min_renda) / (max_renda - min_renda)


df.head()
```

#### 2.2.3 Conversão de variáveis categóricas para números
Aplicamos 3 estratégias distintas conforme a natureza de cada variável:

- **Variável target**: `Status` convertida para binário (1 = Evadido, 0 = demais);
- **Mapeamento ordinal** para variáveis com ordem natural (Faixa Etária);
- **One-Hot Encoding (OHE)** para variáveis sem hierarquia (Gênero, 
Estado Civil, Escola Ensino Médio, Status, Trabalha Atualmente), 
evitando que o modelo interprete uma ordem inexistente entre as categorias.

```python
# Mapeamento ordinal (variáveis com ordem natural)
mapeamento_ordinal = {'Faixa Etária': {'Jovem': 1, 'Adulto': 2, 'Maduro': 3}}


df_convertido = df[['Idade', 'Gênero', 'Estado Civil', 'Renda Familiar (R$)', 'Escola Ensino Médio', 'Trabalha Atualmente', 'Último Período Cursado', 'Desempenho', 'Faixa Etária']].copy()

for col, mapa in mapeamento_ordinal.items():
    df_convertido[col] = df_convertido[col].map(mapa)


# One-Hot Encoding (variáveis sem ordem natural)
colunas_ohe = ['Gênero', 'Estado Civil', 'Escola Ensino Médio', 'Trabalha Atualmente']
df_convertido = pd.get_dummies(df_convertido, columns = colunas_ohe, dtype = int)


# Definindo variável target (o que queremos prever)
df_convertido['Target'] = (df['Status'] == 'Evadido').astype(int)

df_convertido.head()
```
